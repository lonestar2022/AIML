```python
grp_df.head()
```

```python
#resetting the index to get the movie id as column
grp_df.reset_index(inplace=True)
```

```python
grp_df.head()
```

#### (b). selecting each series from grp_df and recreating new df, as rating is a dataframe with size and mean column

```python
mv_id,count,avg= grp_df['movie id'],grp_df['rating']['size'],grp_df['rating']['mean']
```

```python
agg_df=pd.concat([mv_id,count,avg], axis=1)
```

```python
agg_df.head()
```

#### (c). Selecting movies that have more than 100 ratings and sorting it by mean and selecting top 25 records

```python
top25_df=agg_df[agg_df['size'] >100].sort_values(by=['mean'],ascending=False).head(25)
```
